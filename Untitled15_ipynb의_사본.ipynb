```python
#1.계산기라는 기계를 만듭니다.
def 계산기():
    #숫자 2개와 기호 1개를 상자에 담습니다.(.strip()은 빈칸 지우개에요!)
    첫번째_숫자 = float(input("첫번째 숫자:").strip())
    두번째_숫자 = float(input("두번째 숫자:").strip())
    연산_기호 = input("더하기(+) 빼기(-)등 기호선택:").strip()

    #기호가 무엇이냐에 따라 계산을 다르게 합니다.
    if 연산_기호 == '+':
       print(첫번째_숫자 + 두번째_숫자)
    elif 연산_기호 == '-':
       print(첫번째_숫자 - 두번째_숫자)
    elif 연산_기호 == '*':
       print(첫번째_숫자 * 두번째_숫자)
    elif 연산_기호 == '/':
       print(첫번째_숫자 / 두번째_숫자)
    elif 연산_기호 == '%':
       print(첫번째_숫자 % 두번째_숫자)
    elif 연산_기호 == '//':
       print(첫번째_숫자 // 두번째_숫자)

#2 이제 만든 기계로 실제로 작동 시킵니다.
계산기()
```
