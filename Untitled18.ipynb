```python
# 1. 메뉴판 설정 (메뉴명: 가격)
print("--- ☕ GEMINI CAFE 메뉴판 ---")
print("1. 아메리카노 (3,000원)")
print("2. 카페라떼   (3,500원)")
print("3. 바닐라라떼 (4,000원)")
print("4. 카푸치노   (3,800원)")
print("5. 에스프레소 (2,500원)")

# 2. 메뉴 선택
choice = input("\n메뉴 번호를 선택하세요 (1~5): ")

# 3. 선택된 번호에 따라 메뉴와 가격 설정
if choice == "1":
    menu = "아메리카노"; price = 3000
elif choice == "2":
    menu = "카페라떼"; price = 3500
elif choice == "3":
    menu = "바닐라라떼"; price = 4000
elif choice == "4":
    menu = "카푸치노"; price = 3800
elif choice == "5":
    menu = "에스프레소"; price = 2500
else:
    print("❌ 잘못된 번호입니다."); exit() # 프로그램 종료

# 4. 샷 추가 옵션 (모든 메뉴 공통)
add_shot = input(f"> {menu}에 샷을 추가하시겠습니까? (1: 예 / 2: 아니오): ")

if add_shot == "1":
    price += 500
    menu = f"{menu}(샷 추가)"

# 5. 최종 출력 (f-string 사용)
print("-" * 30)
print(f"✅ 주문하신 {menu} 한 잔 준비해 드릴게요.")
print(f"💰 결제 금액은 {price}원입니다.")
print("-" * 30)
```
